```python
# Import Libraries
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
# Load Dataset
dataset = pd.read_csv("youtube.csv")
print("Dataset Preview:")
print(dataset.head())
```

```python
dataset['target'] = np.where(dataset['views'] > dataset['views'].median(), 1, 0)
X = dataset[['likes','dislikes','comment_count','category_id']].values
Y = dataset['target'].values
print("\nX Sample:\n", X[:5])
print("\nY Sample:\n", Y[:5]) 
```

```python
print("\nMissing Values:\n")
print(dataset.isnull().sum())
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
X = imputer.fit_transform(X)
print("\nAfter Imputation (First 5 rows):\n", X[:5])
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_X = LabelEncoder()
X[:, 3] = labelencoder_X.fit_transform(X[:, 3])
print("\nAfter Label Encoding:\n", X[:5])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(
    transformers=[("encoder", OneHotEncoder(), [3])],
    remainder='passthrough'
)
X = ct.fit_transform(X)
print("\nAfter One Hot Encoding:")
print("Shape:", X.shape)
```

```python
labelencoder_Y = LabelEncoder()
Y = labelencoder_Y.fit_transform(Y)
print("\nEncoded Y Sample:\n", Y[:5])
```

```python

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=0
)
print("\nTraining Shape:", X_train.shape)
print("Testing Shape:", X_test.shape)
```

```python
X_train = X_train.toarray()
X_test = X_test.toarray()
print("\nConverted to Dense Matrix")
print("X_train Shape:", X_train.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
print("\nAfter Scaling (First 5 rows):\n", X_train[:5])
```
